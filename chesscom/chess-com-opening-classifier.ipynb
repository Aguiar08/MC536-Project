```python
import re
import numpy as np
import pandas as pd
```

```python
# funcao utilizada para encontrar maior sequencia de movimentos nas openings que configuram
# alguma opening pertencente a database de openings utilizada
# (otimizavel eu tnh qse ctz)

def encontraOpening(moves, openings):
    maiorOpening = 0
    openingName = ''
    for i in range(len(openings)):
        moveString = openings['pgn'][i]
        if moveString in moves and len(moveString)>maiorOpening:
            maiorOpening = len(moveString)
            openingName = openings['name'][i]

    return openingName
```

```python
filename = 'games_db_65_74.csv'

players = pd.read_csv('games/'+filename)

players.head()
```

```python
openings = pd.read_csv('../Openings/openings.csv')

openings.head()
```

```python
players['opening'] = players.apply(lambda x: encontraOpening(x['moves'], openings), axis=1)

players
```

```python
players.to_csv('gamesWithOpenings/'+filename,index = False)
```
