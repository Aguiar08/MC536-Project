```python
import re
import numpy as np
import pandas as pd
```

```python

def encontraOpening(moves, openings):
    maiorOpening = 0
    openingName = ''
    for i in range(len(openings)):
        moveString = openings['pgn'][i]
        if moveString in moves and len(moveString)>maiorOpening:
            maiorOpening = len(moveString)
            openingName = openings['name'][i]

    return openingName
```

```python
filename = 'games_db_1_3.csv'

players = pd.read_csv('games/'+filename)

players.head()
```

```python
players['moves']
```

```python
openings = pd.read_csv('../Openings/openings.csv')

openings.head(10)
```

```python
players['opening'] = players.apply(lambda x: encontraOpening(x['moves'], openings), axis=1)

players.head(10)
```
