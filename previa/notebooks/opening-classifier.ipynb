```python
import numpy as np
import pandas as pd
```

```python
def encontraOpening(moves, openings):
    if type(moves) == float:
        return ''
    for i in range(0,len(openings)):
        moveString = openings['pgn'][i]
        if moveString in moves:
            return openings['name'][i]
    return ''
```

```python
players = pd.read_csv('../data/raw/all_games_lichess.csv')
```

```python
openings = pd.read_csv('../data/processed/openings.csv')
```

```python
players['opening'] = players.apply(lambda x: encontraOpening(x['moves'], openings), axis=1)
players.head(5)
```
